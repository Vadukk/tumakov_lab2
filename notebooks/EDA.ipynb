# Data preprocessing

Заменим все категориальные колонки на численные значения с помощью LabelEncoder, тк мы все равно будем все передавать в деревянную модель

```python
cols_categorial = []
```

```python
df = df.fillna('No')
```

Некоторые численные колонки - это уже закодированные категориальные признаки. Например `MSSubClass`: Identifies the type of dwelling involved in the sale

```python
cols_categorial.append('MSSubClass')
```

```python
encoders_path = '../models/Encoders/'
```

```python
for col in df.columns:
    if isinstance(df.loc[0, col], str):
        cols_categorial.append(col)
        encoder = LabelEncoder()
        
        df[col] = encoder.fit_transform(df[col])
        
        np.save(os.path.join(encoders_path, f'{col}_encoder.npy'), encoder.classes_)
```

```python
df.sample(5)
```

```python
df.drop(columns='Id', inplace=True)
```

Заменим наны в остальных колонках на нули

```python
for col in df.columns:
    if 'No' in list(df[col]):
        df.loc[df[col] == 'No', col] = 0
```

```python
df.to_csv('../data/processed/train.csv', index=False)
```
